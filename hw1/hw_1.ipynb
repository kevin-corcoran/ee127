(a) Let $x,y\in\mathbb{R}^n$ be two unit-norm vectors, that is, such that $\|x\|_2 = \|y\|_2=1$. Show algebraically that the vectors $x-y$ and $x+y$ are orthogonal.
Then, show this graphically by drawing the two vectors on the 2D plane, as well as any other necessary shapes. You may use right angles, circles and straight lines to
make your point.


**Algebraically:**  
**Goal**: show inner product $ \left<x-y, x+y\right> = 0 $

$$
\begin{align}
\left<x-y, x+y \right> & = \left< x-y, x\right> + \left< x-y, y \right> \\
& = \left< x, x \right> - \left< y, x \right> + \left< x , y \right> - \left< y, y \right> \\
& = \left< x, x \right> - \left< x, y \right> + \left< x , y \right> - \left< y, y \right> && && \text{ since $ \left<y, x \right> =  \overline{\left< x, y \right>} = \left< x, y \right> $} \\
& = \|x\|^2_2 - \|y\|^2_2 = 1 - 1 \\
& = 0
\end{align}
$$

proving that $x-y$ and $x+y$ are orthogonal

(b) Show that the following inequalities hold for any vector $x \in \mathbb{R}^{n}$:
$$
   \frac{1}{\sqrt{n}}\|x\|_2 \leq \|x\|_\infty \leq  \|x\|_2 \leq \|x\|_1 \leq \sqrt{n} \|x\|_2 \le n\|x\|_\infty.
$$
   
Hint: For $\|x\|_1\leq \sqrt{n}\|x\|_2$, how might you express $\|x\|_1$ as the dot product of two vectors? Can you then use the Cauchy-Schwarz inequality to
bound this dot product?

$ \frac{1}{\sqrt{n}}\|x\|_2 \leq \|x\|_\infty $  

$$
\begin{align}
\frac{1}{\sqrt{n}}\|x\|_2  = \frac{1}{\sqrt{n}} \sqrt{x_1^2 + \dots +  x_n^2} & \leq \frac{1}{\sqrt{n}} \sqrt{x_k^2 + \dots +  x_k^2}  
        && \text{ where $\max\limits_i|x_i| = x_k \geq x_i$ for $ i = 1, \dots, n$ } \\
& = \frac{\sqrt{n}}{\sqrt{n}} \sqrt{x_k^2} = x_k = \max\limits_i|x_i| \\
& = \|x\|_\infty
\end{align}
$$

$\|x\|_\infty \leq  \|x\|_2$

$$
\begin{align}
\|x\|_\infty = \max\limits_i|x_i| = \sqrt{x_k^2} \leq  \sqrt{x_1^2 + \dots + x_k^2 + \dots + x_n^2} = \|x\|_2
\end{align}
$$

$\|x\|_2 \leq \|x\|_1$ 

Graphically in $\mathbb{R}^2$, $\|x\|_2$ is the hypotnuse of a triangle and $\|x\|_1$ is the sum of the lengths of the sides. Intuitively the inequality holds.
But it's also true by the triangle inequality..

Now assume this is is true for $\mathbb{R}^{n-1}$

**Goal:** show true for $\mathbb{R}^n$

$$
\begin{align}
\|x\|_2^2 & = (x_1^2 + \dots + x_{n-1}^2) + x_n^2 \\
    & \leq (|x_1| + \dots + |x_{n-1}|) + x_n^2 && && \text{ by assumption} \\
    & \leq \left( (|x_1| + \dots + |x_{n-1}|) + |x_n| \right)^2 \\
    & = \|x\|_1^2 \\
    \therefore  \|x\|_2 \leq \|x\|_1
\end{align}
$$

this proves the inequality by induction and generalizes the triangle inequality in higher dimensions

$\|x\|_1 \leq     \sqrt{n} \|x\|_2$

$$
\begin{align}
\text{let } v = \left( sign(x_1), \dots, sign(x_n) \right) && && \text{where $sign(x_i) = \pm 1$} \\
\implies \|x\|_1 & = \left< x, v\right> \\
    & \leq \|x\|_2 \|v\|_2 && && \text{by Cauchy-Schwartz} \\
    & = \sqrt{n} \|x\|_2
\end{align}
$$

$\sqrt{n} \|x\|_2 \le n\|x\|_\infty$

$$
\begin{align}
\sqrt{n} \|x\|_2 & = \sqrt{n} \sqrt{x_1^2 + \dots + x_n^2} \\
    & \leq \sqrt{n} \sqrt{x_k^2 + \dots +  x_k^2}  
        && \text{ where $\max\limits_i|x_i| = x_k \geq x_i$ for $ i = 1, \dots, n$ } \\
    & = n \|x\|_\infty
\end{align}
$$

**2. Gradients and Hessians**  
Compute the Gradients and Hessians for the following functions:  

(a) $g(x) = y^T A x$

**Gradient**

$$
\begin{align}
\text{let $A_{n \times m}$, $x_{m \times 1}$, and $y^T_{1 \times n}$} \\ 
\implies y^T A x  & = y^T \sum_{j=1}^{m} a_{ij}x_j \\
    & = \sum_{i = 1}^{n} y_i \sum_{j=1}^{m} a_{ij}x_j \\
    & = \sum_{i = 1}^{n} \sum_{j=1}^{m} a_{ij}y_i x_j && && \text{ Since $y_i(a_{j1}x_1 + \dots + ) \\  = a_{j1}y_ix_1 + \dots + $} \\
    \\
    \implies \nabla y^T A x & = 
    \begin{pmatrix}
        \frac{\partial{ \sum_{j=1}^{m} a_{ij}y_i x_j  }}{\partial{x_j}}
    \end{pmatrix} && && \text{where this is a column vector in j}\\
    & = \boxed{y^T A}
\end{align}
$$

**Hessian**

$$
\begin{align}
\left[ \nabla^2 g(x) \right]_{ij} & = \frac{ \partial^2{ \sum_{j=1}^{m} a_{ij}y_i x_j } }{ \partial x_i \partial x_j } \\
    & = \frac{\partial }{\partial x_i} \left( \frac{\partial{ \sum_{j=1}^{m} a_{ij}y_i x_j  }}{\partial{x_j}} \right) \\
    & = \frac{\partial }{\partial x_i}  \sum_{j=1}^{m} a_{ij}y_i  \\
    & = \boxed{0}
\end{align}
$$

(a) $g(x_1,x_2) = \dfrac{x_1^2}{4} + \dfrac{x_2^2}{9}$


**First order taylor**
$$
\begin{align}
g(x_1, x_2) & = (1/4 + 1/9)+ 2/4(x_1 -1) + 2/9(x_2-1) \\
    & = 13/36 + 1/2(x_1-1) + 2/9(x_2-1)
\end{align}
$$


**Second order taylor**
$$
\begin{align}
g(x_1, x_2) & = (1/4 + 1/9)+ 2/4(x_1 -1) + 2/9(x_2-1) + 2/(4*2) (x_1-1)^2 + 2/(9*2) (x_2-1)^2 \\
    & = 13/36 + 1/2(x_1-1) + 2/9(x_2-1) + 1/4(x_1-1)^2 + 1/9(x_2-1)^2
\end{align}
$$

**Gradient**

$$
\begin{align}
\nabla g(x_1, x_2) =
        \begin{pmatrix} 
            \frac{x_1}{2} \\
            \frac{2x_2}{9}
        \end{pmatrix}
\end{align}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def g(x, y):
    return x**2/4 + y**2/9
def g1(x, y):
    return 13/36 + 1/2*(x - 1) + 2/9*(y - 1)
def g2(x, y):
    return g1(x, y) + 1/4*(x-1)**2 + 1/9*(y-1)**2
def gradg(x, y):
    return [x/2, 2*y/9]

x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 40)

X, Y = np.meshgrid(x, y)
Z = g(X, Y)

plt.contour(X, Y, Z, colors='black');
grad = gradg(X,Y)
plt.quiver(X, Y, grad[0], grad[1])
```

```python
# plot of first order taylor
Z = g1(X, Y)
plt.contour(X, Y, Z, colors='black');
```

```python
# plot of second order taylor
Z = g2(X, Y)
plt.contour(X, Y, Z, colors='black');
```

(b) $g(x_1,x_2) = x_1 x_2$

**First order taylor**
$$
\begin{align}
g(x_1, x_2) & = 1 + (x_1-1) + (x_2-1)
\end{align}
$$


**Second order taylor**
$$
\begin{align}
g(x_1, x_2) & = 1 + (x_1-1) + (x_2-1) + (x_1-1)(x_2-1)
\end{align}
$$

**Gradient**

$$
\begin{align}
\nabla g(x_1, x_2) =
        \begin{pmatrix} 
            x_2 \\
            x_1
        \end{pmatrix}
\end{align}
$$

```python
def g(x, y):
    return x*y
def g1(x, y):
    return 1 + (x - 1) + (y - 1)
def g2(x, y):
    return g1(x, y) + (x-1)*(y-1)
def gradg(x, y):
    return [y, x]

x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 40)

X, Y = np.meshgrid(x, y)
Z = g(X, Y)

plt.contour(X, Y, Z, colors='black');
grad = gradg(X,Y)
plt.quiver(X, Y, grad[0], grad[1])
```

```python
# plot of first order taylor
Z = g1(X, Y)
plt.contour(X, Y, Z, colors='black');
```

```python
# plot of second order taylor
Z = g2(X, Y)
plt.contour(X, Y, Z, colors='black');
```

As we can see the first order taylors approximations are very poor, but the second order appears to be exact, and in fact it is. If we tried to take higher orders, we would see the terms are zero.
