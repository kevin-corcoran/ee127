# The pendulum

Let consider the following system: $\ddot{y} = -\omega^2 y $. 
Let $X = \begin{bmatrix}
    y \\ \dot{y}
\end{bmatrix}$.
We have that 
$\dot{X} = \begin{bmatrix}
    \dot{y} \\ \ddot{y}
\end{bmatrix}$.

Then $\dot{X} = AX$ is the system we want to solve

### 3.i. Give a matrix such that $\dot{X}=AX$

```python
w = 5
# A = # TO DO
A = np.array([[0, 1],
              [-w**2, 0]])

X0 = np.array([y0, doty0])

# diagonalization
values, vectors = np.linalg.eig(A)
P = vectors
Pinv = np.linalg.inv(P)

# n = 5
# t = 2
# print(values)
# print(np.diag(values))
# print(np.eye(2,2) + np.diag(values*t/n))
```

### 3.l. Write $  X = P\ \begin{bmatrix}
        1 + \lambda_1 \frac{t}{n}    &0\\
        0      &1 + \lambda_2 \frac{t}{n}
    \end{bmatrix}^{\ n} P^{-1} X_0$ 
    
In the notebook, $X_0=X0$, $P = P$, $P^{-1}=Pinv$, $[\lambda_1,\lambda_2] = values$

```python
y0 = 1
doty0 = 0
for j in [0, 1, 2, 3, 4]:
    n = 10**j

    print("n= " + str(n))
    
    X0 = np.array([y0, doty0])

    # diagonalization
    values, vectors = np.linalg.eig(A)
    P = vectors
    Pinv = np.linalg.inv(P)

    # finding X
    nb_points = 1000
    xx = np.zeros(nb_points)
    yy = np.zeros(nb_points)
    for i in range(nb_points):
        t = (10*np.pi*i)/(nb_points*w)
#       X = # TO DO
        D = np.eye(2,2) + np.diag(values*t/n)
        X = np.dot(np.dot(np.dot(P, np.linalg.matrix_power(D,n)), Pinv), X0)
        xx[i] = t
        yy[i] = np.real(X)[0]
    
    y = np.cos(w * xx-0.2)
    plt.plot(xx, yy, 'm')
    plt.plot(xx, y, 'c')
    plt.show()
```
