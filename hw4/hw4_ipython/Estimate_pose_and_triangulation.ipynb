```python
# 1.3
# You complete this ------------------------------------------------------->
def recover_translation(E):
    # E = Essential matrix (3x3 matrix)
    # return t1, t2 , (3x1 arrays)
    
    u, s, vh = np.linalg.svd(E, full_matrices=True)
    # count how many singular values are zero
    count = 0
    for i in s: # note: better to loop backwards and break at the first none zero
        if i < 10**(-2): # singular value is zero
            count+=1
            
    t1 = u[:, len(s)-count:len(s)]
    t2 = -t1
    
    return t1, t2
```

```python
# 1.4
# You complete this -------------------------------------------------------->
def recover_rotation(E):
    # E = Essential matrix (3x3 matrix)
    # return R1, R2, (3x3 matrices)
    
    Rn = np.array(((0,-1,0), (1, 0, 0), (0, 0, 1)))
    Rp = np.array(((0,1,0), (-1, 0, 0), (0, 0, 1)))
    
    u, s, vh = np.linalg.svd(E, full_matrices=True)
    
    Rpp = +u@Rp@vh
    Rnn = -u@Rp@vh
    Rpn = +u@Rn@vh
    Rnp = -u@Rp@vh
    
    a = [Rpp, Rnn, Rpn, Rnp]
    b = [np.linalg.det(Rpp), np.linalg.det(Rnn), np.linalg.det(Rpn), np.linalg.det(Rnp)]
    
    sol = ()
    for i in range(len(b)):
        if b[i] > 0:
            sol = sol + (a[i],)
    
    # make sure det(R) = 1
        
    return sol
#     return R1, R2
    
```

```python
# 1.5
# You complete this -------------------------------------------------------------------->
# helper function to triangulate for individual points
def triangulate_point(P0, P1, pt0, pt1):
    # P0 = K[I, 0] 3x4 matrix
    # P1 = K[R,t] 3x4 matrix
    # pt0 = [x_0,y_0] 2x1 array
    # pt1 = [x_1, y_1] 2x1 array 
    # return three_d_point 1x3 array
 
    # do triangulation for single point
    # create Matrix A, in equation Ax = 0
    A = np.array([pt0[0][0]*P0[2,:] - P0[0,:],
                  pt0[1][0]*P0[2,:] - P0[1,:],
                  pt1[0][0]*P1[2,:] - P1[0,:],
                  pt1[1][0]*P1[2,:] - P1[1,:] ])

    # solve linear equation using svd
    u, s, vh = np.linalg.svd(np.transpose(A), full_matrices=True)
    # count how many singular values are zero
    count = 0
    for i in s: # note: better to loop backwards and break at the first none zero
        if i < 1**(-2): # singular value is zero
            count+=1
            
    # get 3D point by normalizing (dividing by 4th component)
    
    t1 = u[:, len(s)-1-count:len(s)-1] 
    four_d_point = t1/t1[3] # four d point
#     three_d_point = np.transpose(four_d_point[0:3])
    three_d_point = four_d_point[0:3]

#     pdb.set_trace()
    return three_d_point #.flatten()
```

```python
def triangulate(K1, K2, R,t,points0, points1):
    # K1 = Camera Matrix 1 (3x3 matrix)
    # K2 = Camera Matrix 2 (3x3 matrix)
    # R = Rotation Matrix (3x3 matrix)
    # t = Translation vector (3x1 Array)
    # points0 = Nx2 Array of candidate points (pixel coordinates) in imgA (view 0)
    # point1 = Nx2 Array of matches of points1 in imgB (view 1)
    
    # create projection Matrices from R,t, K1, K2
    P0 = np.dot(K1[0:3, 0:3], np.eye(3,4))
    P1 = np.dot(K1[0:3,0:3], np.hstack((R,t)))
    
    
    res = [triangulate_point(P0,P1,x.reshape((2,1)),y.reshape((2,1))) for x,y in zip(points0, points1)]
#     pdb.set_trace()
    res = np.squeeze(np.asarray(res))
        
    
    return res
```

```python
# 1.6
# You complete this ----------------------------------------------------------------->
def decompose_essential_matrix(E,K1, K2, points0, points1):
    # E = Essential matrix (3x3 matrix)
    # K1 = Camera Matrix 1 (3x3 matrix) 
    # K2 = Camera Matrix 2 (3x3 matrix) 
    # points0 = Nx2 Array of candidate points (pixel coordinates) in imgA (view 0)
    # points1 = Nx2 Array of matches of points1 in imgB (view 1)
    
    
    # get candidate R, and t
    R1, R2 = recover_rotation(E)
    t1, t2 = recover_translation(E)
    t1 = t1.reshape((3,1))
    t2 = t2.reshape((3,1))
    
    # get all possible combinations
    transformation_combinations = [(R1, t1), (R1, t2), (R2,t1), (R2, t2)]
    positive_count = []
    
    # choose the combination with most points in front of both cameras
    for transformation in transformation_combinations:
        R_tmp = transformation[0]
        t_tmp = transformation[1]
        three_d_points = triangulate(K1, K2, R_tmp, t_tmp, points0, points1)
        n_positive = len(np.where(three_d_points[2,:] > 0)[0])
        positive_count.append(n_positive)
        
    best_idx = np.argmax(positive_count)
    # get best transformation, complete this portion---------------------------------->
    best_transformation = transformation_combinations[best_idx]

 
    return best_transformation
```
